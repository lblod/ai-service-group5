```python
def ask_llm(text):
    response = llm.create_chat_completion(
    messages = [
        {"role": "assistant", 
        "content": f"""You are an assistant that will help label a text. The text you will be given comes from Flemish government notes. 
                    For the content you are given, you need to classify according to these 3 overall criteria: Thema, Doelgroep, and Type. For each criteria, 
                    there is a list of labels you need to refer to label the text. For Thema, use {thema}. For Doelgroep, refer to {doelgroep}. For Type, refer to {type}. 
                    Each text needs to have corresponding labels from all 3 criteria. For the Doelgroep criteria, there can be multiple labels. Return the criteria and the labels 
                    associated for the text given."""            
        },
        {
            "role":"user",
            "content":f"{text}"
        }
    ]
    )
    return response['choices'][0]['message']['content']

# Test 
print(ask_llm(text))
```
